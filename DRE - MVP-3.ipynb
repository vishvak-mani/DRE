```python
import ipywidgets as widgets
from ipywidgets import HBox, VBox
from IPython.display import display
from ipywidgets import IntProgress
from ipywidgets import interact, interact_manual
import pandas as pd 
import time

metadata = pd.read_csv('investor_activity.csv')
metadata
```

```python
output = widgets.Output()

ALL = 'All regions'
def unique_sorted_values_plus_ALL(array):
    unique = array.unique().tolist()
    unique.sort()
    unique.insert(0, ALL)
    return unique

dropdown_region = widgets.Dropdown(options = unique_sorted_values_plus_ALL(metadata.region))

def dropdown_region_eventhandler(change):
    if (change.new == ALL):
        display(metadata)
    else:
        display(metadata[metadata.region == change.new])
        
dropdown_region.observe(dropdown_region_eventhandler, names='value')

ALL = 'All industries'
def unique_sorted_values_plus_ALL(array):
    unique = array.unique().tolist()
    unique.sort()
    unique.insert(0, ALL)
    return unique

dropdown_industry = widgets.Dropdown(options = unique_sorted_values_plus_ALL(metadata.industry))

def dropdown_industry_eventhandler(change):
    if (change.new == ALL):
        display(metadata)
    else:
        display(metadata[metadata.industry == change.new])
        
dropdown_industry.observe(dropdown_industry_eventhandler, names='value')

#display(dropdown_region)
#display(dropdown_industry)

def common_filtering(region, industry):
    output.clear_output()
    
    if (region == ALL) & (industry == ALL):
        common_filter = metadata
    elif (region == ALL):
        common_filter = metadata[metadata.industry == industry]
    elif (industry == ALL):
        common_filter = metadata[metadata.region == region]
    else:
        common_filter = metadata[(metadata.region == region) & 
                                  (metadata.industry == industry)]
    
    with output:
        display(common_filter)
        
def dropdown_region_eventhandler(change):
    common_filtering(dropdown_industry.value, change.new)
    
def dropdown_industry_eventhandler(change):
    common_filtering(change.new, dropdown_region.value)
    
dropdown_region.observe(dropdown_region_eventhandler, names='value')
dropdown_industry.observe(dropdown_industry_eventhandler, names='value')

input_widgets = widgets.HBox([dropdown_region, dropdown_industry])
#display(input_widgets)

df = pd.read_csv('mock_deal_database_1-2.csv')
df = df.rename(columns={'Industry Type': 'Industry_Type', 'Issuance Type': 'Issuance_Type'})
```

```python
output = widgets.Output()

ALL = 'All issuance'
def unique_sorted_values_plus_ALL(array):
    unique = array.unique().tolist()
    unique.sort()
    unique.insert(0, ALL)
    return unique

dropdown_Issuance_Type = widgets.Dropdown(options = unique_sorted_values_plus_ALL(df.Issuance_Type))

def dropdown_Issuance_Type_eventhandler(change):
    if (change.new == ALL):
        display(df)
    else:
        display(df[df.Issuance_Type == change.new])
        
dropdown_Issuance_Type.observe(dropdown_Issuance_Type_eventhandler, names='value')

ALL = 'All industries'
def unique_sorted_values_plus_ALL(array):
    unique = array.unique().tolist()
    unique.sort()
    unique.insert(0, ALL)
    return unique

dropdown_Industry_Type = widgets.Dropdown(options = unique_sorted_values_plus_ALL(df.Industry_Type))

def dropdown_Industry_Type_eventhandler(change):
    if (change.new == ALL):
        display(df)
    else:
        display(df[df.Industry_Type == change.new])
        
dropdown_Industry_Type.observe(dropdown_Industry_Type_eventhandler, names='value')

#display(dropdown_region)
#display(dropdown_industry)

def common_filtering(Issuance_Type, Industry_Type):
    
    output.clear_output()
    
    if (Issuance_Type == ALL) & (Industry_Type == ALL):
        common_filter = df
    elif (Issuance_Type == ALL):
        common_filter = df[df.Industry_Type == Industry_Type]
    elif (Industry_Type == ALL):
        common_filter = df[df.Issuance_Type == Issuance_Type]
    else:
        common_filter = df[(df.Issuance_Type == Issuance_Type) & 
                                  (df.Industry_Type == Industry_Type)]
    
    with output:
        display(common_filter)
        
def dropdown_Issuance_Type_eventhandler(change):
    common_filtering(dropdown_Industry_Type.value, change.new)
    
def dropdown_Industry_Type_eventhandler(change):
    common_filtering(change.new, dropdown_Issuance_Type.value)
    
dropdown_Issuance_Type.observe(dropdown_Issuance_Type_eventhandler, names='value')
dropdown_Industry_Type.observe(dropdown_Industry_Type_eventhandler, names='value')

Input_widgets = widgets.HBox([dropdown_Issuance_Type, dropdown_Industry_Type])
#display(Input_widgets)
```

```python
def normalize(amount_text):
    d = {
        'K': 1000,
        'M': 1000000,
        'B': 1000000000
    }
    if amount_text[-1] in d:
        # separate out the K, M, or B
        num, magnitude = amount_text[:-1], amount_text[-1]
        return int(float(num[1:]) * d[magnitude])
    else:
        return float(amount_text[1:])
    
df['Target Raise'] = df['Target Raise'].apply(lambda x: (normalize(x) if x != "n/a" else x))
df['Amount Raised'] = df['Amount Raised'].apply(lambda x: (normalize(x) if x != "n/a" else x))

print("Toggle your minimum preferred amount:\n")
@interact
def show_articles_more_than(column='Target Raise', Amount=100000000):
    return df.loc[df[column] > Amount]
```

```python
# Parse the stringified features into their corresponding python objects
from ast import literal_eval

features = ['industry', 'issuance_type']
for feature in features:
    metadata[feature] = metadata[feature].apply(literal_eval)
    
# Function to convert all strings to lower case and strip names of spaces
def clean_data(x):
    if isinstance(x, list):
        return [str.lower(i.replace(" ", "")) for i in x]
    else:
        #Check if director exists. If not, return empty string
        if isinstance(x, str):
            return str.lower(x.replace(" ", ""))
        else:
            return ''
        
# Apply clean_data function to your features.
features = ['issuance_type', 'region', 'industry']

for feature in features:
    metadata[feature] = metadata[feature].apply(clean_data)
    
# Function to convert all strings to lower case and strip names of spaces
def clean_data(x):
    if isinstance(x, list):
        return [str.lower(i.replace(" ", "")) for i in x]
    else:
        #Check if director exists. If not, return empty string
        if isinstance(x, str):
            return str.lower(x.replace(" ", ""))
        else:
            return ''
        
# Apply clean_data function to your features.
features = ['issuance_type', 'region', 'industry']

for feature in features:
    metadata[feature] = metadata[feature].apply(clean_data)
    
def create_soup(x):
    return ' '.join(x['region']) + ' ' + ' '.join(x['issuance_type']) + ' ' + ' '.join(x['industry'])

metadata['soup'] = metadata.apply(create_soup, axis=1)

# Import CountVectorizer and create the count matrix
from sklearn.feature_extraction.text import CountVectorizer

count = CountVectorizer(stop_words='english')
count_matrix = count.fit_transform(metadata['soup'])

# Compute the Cosine Similarity matrix based on the count_matrix
from sklearn.metrics.pairwise import cosine_similarity

cosine_sim = cosine_similarity(count_matrix, count_matrix)

# Reset index of your main DataFrame and construct reverse mapping as before
metadata = metadata.reset_index()
indices = pd.Series(metadata.index, index=metadata['investor'])

def get_recommendations(title, cosine_sim=cosine_sim):
    # Get the index of the movie that matches the title
    idx = indices[title]

    # Get the pairwsie similarity scores of all movies with that movie
    sim_scores = list(enumerate(cosine_sim[idx]))

    # Sort the movies based on the similarity scores
    sim_scores = sorted(sim_scores, key=lambda x: x[1], reverse=True)

    # Get the scores of the 10 most similar movies
    sim_scores = sim_scores[1:11]

    # Get the movie indices
    movie_indices = [i[0] for i in sim_scores]

    # Return the top 10 most similar movies
    return ((metadata['investor'])).iloc[movie_indices]

# Reset index of your main DataFrame and construct reverse mapping as before
metadata = metadata.reset_index()
indices = pd.Series(metadata.index, index=metadata['investor'])

inv = input("Invited by (Admin/RRep/Issuer): ")
pref = input("Enter your preference: ")

max_count = 100

f = IntProgress(min=0, max=max_count) 
display(f) 

count = 0
while count <= max_count:
    f.value += 7 
    time.sleep(.1)
    count += 7
```

```python
bttn = widgets.Button(description='Similar Investors', button_style='primary')

def bttn_eventhandler(obj):
    print("Similar investors to" + ' ' + pref + ": \n")
    Placeholder = get_recommendations(pref, cosine_sim)
    placeholder = list(Placeholder)
    print(placeholder)

bttn.on_click(bttn_eventhandler)

btn1 = widgets.Button(description='Most Popular Deals', button_style='success')
btn = widgets.Button(description='Least Popular Deal', button_style='danger')
btn2 = widgets.Button(description='Viewed by Similar', button_style='warning')
btn3 = widgets.Button(description='Invested by Similar', button_style='info')
btn4 = widgets.Button(description='Invited by', button_style='primary')

def btn_eventhandler(obj):
    activity_ = pd.read_csv('Investor_Actions_Invested.csv')
    Hm = activity_['deals_invested'].value_counts().idxmin()
    print("Deal:" + '' + Hm)
    print("\n")
    print("No. of times invested:")
    ph = activity_['deals_invested'].value_counts().min()
    print(ph)
    print("\n.........................\n")
    
def btn4_eventhandler(obj):
    print("Invited by:" + ' ' + inv)
    print("\n.........................\n")
    
def btn1_eventhandler(obj):
    activity_ = pd.read_csv('Investor_Actions_Invested.csv')
    hm = activity_.deals_invested.mode()
    print(hm)
    print("\n")
    print("No. of times invested:")
    Ph = activity_['deals_invested'].value_counts().max()
    print(Ph)
    print("\n.........................\n")
    
def btn2_eventhandler(obj):
    activity = pd.read_csv('Investor_Actions_Viewed.csv')
    print("Deals Viewed by Investors similar to" + ' ' + pref + ": \n")
    Placeholder = get_recommendations(pref, cosine_sim)
    placeholder = list(Placeholder)
    heya = activity.loc[activity['investor'].isin(placeholder)]
    print(heya)
    
def btn3_eventhandler(obj):
    activity_ = pd.read_csv('Investor_Actions_Invested.csv')
    print("Deals Invested by Investors similar to" + ' ' + pref + ": \n")
    Placeholder = get_recommendations(pref, cosine_sim)
    placeholder = list(Placeholder)
    Heya = activity_.loc[activity_['investor'].isin(placeholder)]
    print(Heya)
    
btn.on_click(btn_eventhandler)
btn1.on_click(btn1_eventhandler)
btn2.on_click(btn2_eventhandler)
btn3.on_click(btn3_eventhandler)
btn4.on_click(btn4_eventhandler)


INput_widgets = widgets.HBox([btn1, btn, btn4])
INNput_widgets = widgets.HBox([bttn, btn2, btn3])
```
